# HDBSCAN: basic demo

```python
import hdbscan
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = np.load('data/input/clusterable_data.npy')
```

```python
data.shape
```

```python
data
```

```python
plt.scatter(*data.T, s=50, linewidth=0, c='b', alpha=0.25)
```

```python
clusterer = hdbscan.HDBSCAN(min_cluster_size=15).fit(data)
```

```python
color_palette = sns.color_palette('husl', 6)
color_palette
```

```python
cluster_colors = [color_palette[x] if x >= 0
                  else (0.5, 0.5, 0.5)
                  for x in clusterer.labels_]

np.unique(clusterer.labels_)
```

```python
cluster_member_colors = [sns.desaturate(x, p) for x, p in
                         zip(cluster_colors, clusterer.probabilities_)]
```

```python
plt.scatter(*data.T, s=50, linewidth=0, c=cluster_member_colors, alpha=0.25)
```
